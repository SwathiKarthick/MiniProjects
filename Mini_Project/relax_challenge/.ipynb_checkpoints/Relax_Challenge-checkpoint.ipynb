```python
#Import CSV data
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
users = pd.read_csv('takehome_users.csv', encoding='latin-1')
users
```

```python

user_engagement = pd.read_csv('takehome_user_engagement.csv', encoding='latin-1')
user_engagement
```

```python
rows = user_engagement.loc[user_engagement['user_id'] == 2]
print(rows)
```

```python
import datetime as dt
user_engagement['time_stamp'] = pd.to_datetime(user_engagement['time_stamp'])
user_engagement['time_stamp'] = user_engagement['time_stamp'].dt.normalize()
window=7
#delete days that are the same
user_engagement = user_engagement.drop_duplicates()
user_engagement
```

```python
#group by user id
grouped_user_engagement = user_engagement.groupby('user_id').rolling(window='{:d}D'.format(window), 
                                                                     on='time_stamp').sum()
grouped_user_engagement
```

```python
users['adopted'] = 0
```

```python
for user_id, frame in grouped_user_engagement.groupby(level=0):
    #calculate maximum of visited
    if max(grouped_user_engagement.loc[user_id]['visited']) >= 3.0:
        users.loc[users['object_id'] == user_id, 'adopted'] = 1
        
users
```

PUT CLASS BALANCE

```python
users['adopted'].value_counts()
```

```python
users['creation_time'] = pd.to_datetime(users['creation_time'])
users_creationtime_adopted_only = users[['creation_time', 'adopted']]
users_creationtime_adopted_only = users_creationtime_adopted_only.set_index('creation_time')
df_creation_time = \
users_creationtime_adopted_only.groupby([(users_creationtime_adopted_only.index.year),
                                         (users_creationtime_adopted_only.index.month)]).sum()
df_creation_time
```

```python
import matplotlib.pyplot as plt
df_creation_time.plot(figsize=(15,4))
plt.show()
```

```python
from time import mktime

for time in users['creation_time']:
    time = mktime(time.timetuple())
    
users['creation_time'] = [mktime(time_notunix.timetuple()) for time_notunix in users['creation_time']]
users
```

```python
X = users[['creation_time', 'creation_source', 'opted_in_to_mailing_list', 
           'enabled_for_marketing_drip']]
y = users['adopted']
X_dummies = pd.get_dummies(X, drop_first=True)
X_dummies.columns
```

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import recall_score

c_space = np.logspace(-5,8,15)
param_grid = {'C':c_space, 'penalty':['l1','l2']}

X_train, X_test, y_train, y_test= train_test_split(X_dummies, y, test_size=0.3, random_state=42)

logreg_classbalanced = LogisticRegression(class_weight='balanced')

logreg_cv_class_balanced = GridSearchCV(logreg_classbalanced, param_grid, cv=5, scoring='f1')
logreg_cv_class_balanced.fit(X_train, y_train)
```

```python
y_pred1 = logreg_cv_class_balanced.predict(X_test)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, y_pred1)
```

```python
from sklearn.ensemble import RandomForestClassifier

param_grid_rfc = { 
    'n_estimators': [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)],
    'max_features': ['auto', 'sqrt', 'log2']
}

rfc = RandomForestClassifier(random_state=42, class_weight='balanced')
rfc_cv = GridSearchCV(rfc, param_grid_rfc, cv=5, scoring='f1', verbose=2)
rfc_cv.fit(X_train,y_train)
```

```python
y_pred2 = rfc_cv.predict(X_test)
```

```python
f1_score(y_test, y_pred2)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree1 = DecisionTreeClassifier(max_depth=2, random_state=42, class_weight='balanced')
dtree1.fit(X_train,y_train)
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(dtree1, out_file=dot_data, filled=True,
                feature_names = X_train.columns, class_names=['Not adopted', 'Adopted'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

```python
from sklearn.utils import resample

# concatenate our training data back together
X_resample = pd.concat([X_train, y_train], axis=1)

# separate minority and majority classes
not_adopted = X_resample[X_resample['adopted']==0]
adopted = X_resample[X_resample['adopted']==1]

# upsample minority
adopted_upsampled = resample(adopted,
                          replace=True, # sample with replacement
                          n_samples=len(not_adopted), # match number in majority class
                          random_state=42) # reproducible results

upsampled = pd.concat([not_adopted, adopted_upsampled])
upsampled.head()
```

```python
X_train_upsampled = upsampled.drop('adopted', axis=1)
y_train_upsampled = upsampled['adopted']
```

```python
logreg2 = LogisticRegression()

logreg_cv_2 = GridSearchCV(logreg2, param_grid, cv=5, scoring='f1')
logreg_cv_2.fit(X_train_upsampled,y_train_upsampled)
```

```python
y_pred3 = logreg_cv_2.predict(X_test)
f1_score(y_test, y_pred3)
```

```python
rfc_2 = RandomForestClassifier(random_state=42)
rfc_cv_2 = GridSearchCV(rfc_2, param_grid_rfc, cv=5, scoring='f1', verbose=2)
rfc_cv_2.fit(X_train_upsampled,y_train_upsampled)
```

```python
y_pred4 = rfc_cv_2.predict(X_test)
```

```python
f1_score(y_test, y_pred4)
```
